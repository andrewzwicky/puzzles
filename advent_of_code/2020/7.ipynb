```python
def func(data):
    acc=0
    line=0
    
    lines_covered=set()
    
    while True:
        if line in lines_covered:
            break
        lines_covered.add(line)
        op,num = data[line].split(' ')
        num=int(num)
        if op == "acc":
            acc += num
            line += 1
        if op == "jmp":
            line += num
        if op == "nop":
            line += 1
        
    print(acc)
    return acc
```
