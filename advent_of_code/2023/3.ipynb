```python
digits_re = re.compile('\d+')
```

```python
def parse(str_data):
    next_map_i = 0
    output_map = {}
    ret = []
    for line in str_data:
        line_map = {}
        for m in list(digits_re.finditer(line)):
            for _i in range(*m.span()):
                line_map[_i] = int(next_map_i)
            output_map[next_map_i] = int(m.group(0))
            next_map_i += 1

        this_line = [line_map[i] if i in line_map.keys() else x for i,x in enumerate(line)]
        ret.append(this_line)
    return np.array(ret), output_map
```

```python
def calc(data_in):
    big_set = set()
    array, output_map = parse(data_in)
    char_indices = np.argwhere(np.logical_and(array != '.', np.logical_not(np.char.isdigit(array))))
    for this_index in char_indices:
        neighs = array[neighbors(array,this_index,center=False)]
        big_set.update(set(map(int, neighs[neighs != '.'])))
    print(big_set)
    print(output_map)
    print([output_map[i] for i in big_set])
    return sum([output_map[i] for i in big_set])
```
